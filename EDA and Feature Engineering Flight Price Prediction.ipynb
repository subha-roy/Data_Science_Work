```python
# Feature Engineering
df["Date"]=df["date"].str.split("/").str[0]
df["Month"]=df["date"].str.split("/").str[1]
df["Year"]="20"+df["date"].str.split("/").str[2]
```

```python
df.head()
```

```python
#but new columns are in object, so we have to convet it into numeric data
df["Date"]=df["Date"].astype(int)
df["Month"]=df["Month"].astype(int)
df["Year"]=df["Year"].astype(int)
```

```python
df.info()
```

```python
df.drop("date",axis=1,inplace=True)
```

```python
df.head(2)
```

```python
## if there is other str attached with the time 
df["arr_time"]=df["arr_time"].apply(lambda x:x.split(" ")[0])
df["arr_hour"]=df["arr_time"].str.split(":").str[0]
df["arr_min"]=df["arr_time"].str.split(":").str[1]
```

```python
df.head(2)
```

```python
df["arr_hour"]=df["arr_hour"].astype(int)
df["arr_min"]=df["arr_min"].astype(int)
```

```python
df.info()
```

```python
df.drop("arr_time",axis=1,inplace=True)
```

```python
df.head(3)
```

```python
## if there is other str attached with the time 
df["dep_time"]=df["dep_time"].apply(lambda x:x.split(" ")[0])
df["dep_hour"]=df["dep_time"].str.split(":").str[0]
df["dep_min"]=df["dep_time"].str.split(":").str[1]
```

```python
df["dep_hour"]=df["dep_hour"].astype(int)
df["dep_min"]=df["dep_min"].astype(int)
```

```python
df.drop("dep_time",axis=1,inplace=True)
```

```python
df.info()
```

```python
df["stop"] = df["stop"].replace('\n\t\t\t\t\t\t\t\t\t\t\t\t',' ', regex=True)
df["stop"] = df["stop"].replace('Via',' ', regex=True)
```

```python
df['stop'].unique()
```

```python
df["stop"] = df["stop"].replace('   ',' ', regex=True)
```

```python
df['stop'].unique()
```

```python
df["no_stops"] = df["stop"].str.split(" ").str[0]
df["place_stops"] = df["stop"].str.split(" ").str[1]
```

```python
df.head()
```

```python
df.drop("no_stopes",axis=1,inplace=True)
df.drop("place_stopes",axis=1,inplace=True)
```

```python
df["no_stops"].unique()
```

```python
df["place_stops"].unique()
```

```python
df["no_stops"]=df["no_stops"].map({"non-stop":0,"1-stop":1,"2+-stop":3})
```

```python
df.head()
```

```python
df["no_stops"].unique()
```

```python
df["duration_hour"]=df["time_taken"].str.split(" ").str[0].str.split("h").str[0]
df["duration_min"]=df["time_taken"].str.split(" ").str[1].str.split("m").str[0]
```
